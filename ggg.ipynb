```python
import numpy as np

def dot_product(x, y):
    # Convert lists to NumPy arrays
    x = np.array(x)
    y = np.array(y)
    
    # Calculate the dot product using element-wise multiplication and sum
    product = np.sum(x * y)
    return product 

a = [1, 2, 3]
b = [4, 5, 6]

dot_product(a, b)
# Expected Output: 32 (calculated as 1*4 + 2*5 + 3*6)
```

```python
import matplotlib.pyplot as plt

# Data (Customer waiting times)
x = [43.1, 35.6, 37.5, 36.5, 45.3, 43.4,
     40.3, 50.2, 47.3, 31.2, 42.2, 45.5,
     30.3, 31.4, 35.6, 45.2, 54.1, 45.6,
     36.5, 43.1]

# Create figure
plt.figure(figsize=(8, 6))

# Plot the histogram with 6 bins
plt.hist(x, bins=6, color='green', label="Waiting Time Distribution", edgecolor='black')

# Label axes and set title
plt.xlabel("Waiting Time (minutes)")
plt.ylabel("Number of Customers")
plt.title("Customer Waiting Times")
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10

# Create a figure with 3 rows, 4 columns, and set figsize to 15x15
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15, 15))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop to generate and plot y = nx for n in range(1, 13)
for i in range(12):
    y_new = (i + 1) * x  # Compute y_new for each subplot
    axes[i].plot(x, y_new, label=f'y = {(i + 1)}x')
    axes[i].set_title(f'Subplot {i // 4}, {i % 4}')  # Row, Column indexing
    axes[i].set_xlabel('x')
    axes[i].set_ylabel('y')
    axes[i].legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
